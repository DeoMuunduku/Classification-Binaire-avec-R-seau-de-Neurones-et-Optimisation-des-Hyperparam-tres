```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Chargement des données
data_path = 'data.csv'  # Assurez-vous que ce chemin est correct pour votre environnement
data = pd.read_csv(data_path)

# Suppression de la colonne inutile
data.drop(['Unnamed: 32'], axis=1, inplace=True)

# Encodage du diagnostic (M/B)
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])

# Séparation des caractéristiques et de la cible
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

# Équilibrage des données avec SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Normalisation des caractéristiques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Sélection de caractéristiques avec RFE
estimator = LogisticRegression()
selector = RFE(estimator, n_features_to_select=10, step=1)
selector = selector.fit(X_scaled, y_resampled)
X_selected = selector.transform(X_scaled)

# Division en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_resampled, test_size=0.2, random_state=42)

# Fonction pour créer le modèle
def create_model(input_shape, neurons=64, activation='relu', optimizer='adam'):
    model = Sequential([
        Dense(neurons, activation=activation, input_shape=(input_shape,)),
        Dense(int(neurons / 2), activation=activation),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Configuration de GridSearchCV pour l'optimisation des hyperparamètres
model = KerasClassifier(build_fn=create_model, verbose=0, input_shape=X_train.shape[1])
param_grid = {
    'epochs': [5, 800],
    'batch_size': [10, 100],
    'neurons': [32, 128],
    'activation': ['relu', 'tanh'],
    'optimizer': ['adam', 'rmsprop']
}
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train, y_train)

# Affichage des meilleurs paramètres
print("Meilleure précision: %f utilisant %s" % (grid_result.best_score_, grid_result.best_params_))

# Utilisation des meilleurs paramètres pour re-entraîner le modèle
best_params = grid_result.best_params_
model = create_model(
    input_shape=X_train.shape[1],
    neurons=best_params['neurons'],
    activation=best_params['activation'],
    optimizer=best_params['optimizer']
)
history = model.fit(X_train, y_train, epochs=best_params['epochs'], batch_size=best_params['batch_size'], validation_split=0.2, verbose=1)

# Continuation du code précédent...

# Visualisation de la distribution des diagnostics avant et après SMOTE
plt.figure(figsize=(14, 7))

# Distribution des diagnostics avant SMOTE
plt.subplot(1, 2, 1)
plt.hist(data['diagnosis'], bins=np.arange(data['diagnosis'].min(), data['diagnosis'].max() + 1.5) - 0.5, rwidth=0.8)
plt.title('Distribution avant SMOTE')
plt.xlabel('Diagnostic')
plt.ylabel('Nombre')
plt.xticks([0, 1], ['Bénin', 'Malin'])

# Distribution des diagnostics après SMOTE
plt.subplot(1, 2, 2)
plt.hist(y_resampled, bins=np.arange(y_resampled.min(), y_resampled.max() + 1.5) - 0.5, rwidth=0.8)
plt.title('Distribution après SMOTE')
plt.xlabel('Diagnostic')
plt.ylabel('Nombre')
plt.xticks([0, 1], ['Bénin', 'Malin'])

plt.tight_layout()
plt.show()


# Évaluation du modèle et affichage des performances
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'\nPrécision du modèle: {accuracy * 100:.2f}%')

y_pred_proba = model.predict(X_test).ravel()
y_pred = np.where(y_pred_proba > 0.5, 1, 0)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f'\nPrécision (Precision): {precision:.2f}')
print(f'Sensibilité (Recall): {recall:.2f}')
print(f'Score F1: {f1:.2f}')
print(f'Aire sous la courbe ROC (AUC): {roc_auc:.2f}')

# Courbes d'apprentissage
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Précision Entraînement')
plt.plot(history.history['val_accuracy'], label='Précision Validation')
plt.title('Courbe de Précision')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perte Entraînement')
plt.plot(history.history['val_loss'], label='Perte Validation')
plt.title('Courbe de Perte')
plt.legend()
plt.show()

# Courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label=f'ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taux de Faux Positif')
plt.ylabel('Taux de Vrai Positif')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.show()

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title('Matrice de confusion')
plt.xlabel('Prédictions')
plt.ylabel('Véritables')
plt.show()

# Courbe de précision-rappel
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
plt.plot(recall, precision, marker='.')
plt.title('Courbe de Précision-Rappel')
plt.xlabel('Rappel')
plt.ylabel('Précision')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, Dropout, MaxPooling1D, BatchNormalization
from scikeras.wrappers import KerasClassifier

# Charger les données
data_path = 'data.csv'  # Ajustez ce chemin selon votre besoin
data = pd.read_csv(data_path)

# Prétraitement
data = data.drop(['Unnamed: 32'], axis=1)
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

# Équilibrage des données avec SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Normalisation des caractéristiques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Division en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42)

# Adapter les données pour CNN
X_train_cnn = np.expand_dims(X_train, axis=2)
X_test_cnn = np.expand_dims(X_test, axis=2)

# Fonction de création de modèle CNN
def create_cnn_model(filters=64, kernel_size=3, activation='relu', optimizer='adam'):
    model = Sequential([
        Conv1D(filters=filters, kernel_size=kernel_size, activation=activation, input_shape=(X_train_cnn.shape[1], 1)),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(64, activation=activation),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Wrap du modèle Keras pour utiliser avec GridSearchCV
model = KerasClassifier(model=create_cnn_model, verbose=0)

# Grille de paramètres pour GridSearchCV
param_grid = {
    'model__filters': [32, 64],
    'model__kernel_size': [3, 5],
    'model__activation': ['relu', 'tanh'],
    'model__optimizer': ['adam', 'rmsprop'],
    'epochs': [50, 100],
    'batch_size': [16, 32]
}

# GridSearchCV pour trouver les meilleurs hyperparamètres
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train_cnn, y_train)

# Affichage des meilleurs hyperparamètres
print("Meilleure précision: %f utilisant %s" % (grid_result.best_score_, grid_result.best_params_))

# Entraînement du modèle avec les meilleurs paramètres trouvés
best_model = grid_result.best_estimator_
history = best_model.fit(X_train_cnn, y_train, validation_split=0.2)

# Évaluation du modèle
y_pred = best_model.predict(X_test_cnn)
y_pred_proba = best_model.predict_proba(X_test_cnn)[:, 1]

# Calcul des métriques de performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Affichage des métriques de performance
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
print(f'ROC AUC: {roc_auc:.2f}')

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Matrice de confusion')
plt.ylabel('Valeurs Réelles')
plt.xlabel('Valeurs Prédites')
plt.show()

# Courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc_value = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Courbe ROC (aire = {roc_auc_value:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de Faux Positif')
plt.ylabel('Taux de Vrai Positif')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Chargement des données
data_path = 'data.csv'  # Remplacez par le chemin de votre fichier
data = pd.read_csv(data_path)

# Suppression des colonnes inutiles
if 'Unnamed: 32' in data.columns:
    data.drop(['Unnamed: 32'], axis=1, inplace=True)

# Encodage des étiquettes
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])

# Séparation en caractéristiques et cible
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

# Équilibrage avec SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Sélection de caractéristiques
estimator = LogisticRegression()
selector = RFE(estimator, n_features_to_select=10, step=1)
selector = selector.fit(X_scaled, y_resampled)
X_selected = selector.transform(X_scaled)

# Division en ensembles d'apprentissage et de test
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_resampled, test_size=0.2, random_state=42)

# Définition du modèle DNN
def create_dense_model(input_shape, neurons=64, activation='relu', optimizer='adam', dropout_rate=0.5):
    model = Sequential([
        Dense(neurons, activation=activation, input_shape=(input_shape,)),
        Dropout(dropout_rate),
        Dense(int(neurons / 2), activation=activation),
        Dropout(dropout_rate / 2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Configuration de GridSearchCV
model = KerasClassifier(build_fn=create_dense_model, input_shape=X_train.shape[1], verbose=0)
param_grid = {
    'epochs': [100, 200],  # Ajustez ces valeurs selon vos besoins
    'batch_size': [10, 20],
    'neurons': [32, 64],
    'activation': ['relu', 'tanh'],
    'optimizer': ['adam', 'rmsprop'],
    'dropout_rate': [0.5, 0.3]
}
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train, y_train)

# Affichage des meilleurs paramètres
print("Meilleure précision: %f utilisant %s" % (grid_result.best_score_, grid_result.best_params_))

# Re-entraînement du modèle avec les meilleurs paramètres
best_params = grid_result.best_params_
model = create_dense_model(
    input_shape=X_train.shape[1],
    neurons=best_params['neurons'],
    activation=best_params['activation'],
    optimizer=best_params['optimizer'],
    dropout_rate=best_params['dropout_rate']
)
history = model.fit(X_train, y_train, epochs=best_params['epochs'], batch_size=best_params['batch_size'], validation_split=0.2, verbose=1)

# Évaluation du modèle
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'\nPrécision du modèle: {accuracy * 100:.2f}%')

y_pred_proba = model.predict(X_test).ravel()
y_pred = np.where(y_pred_proba > 0.5, 1, 0)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f'\nPrécision (Precision): {precision:.2f}')
print(f'Sensibilité (Recall): {recall:.2f}')
print(f'Score F1: {f1:.2f}')
print(f'Aire sous la courbe ROC (AUC): {roc_auc:.2f}')

# Visualisations
# Courbes d'apprentissage
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Précision Entraînement')
plt.plot(history.history['val_accuracy'], label='Précision Validation')
plt.title('Courbe de Précision')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Perte Entraînement')
plt.plot(history.history['val_loss'], label='Perte Validation')
plt.title('Courbe de Perte')
plt.legend()
plt.show()

# Courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label=f'ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taux de Faux Positif')
plt.ylabel('Taux de Vrai Positif')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.show()

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title('Matrice de confusion')
plt.xlabel('Prédictions')
plt.ylabel('Véritables')
plt.show()

# Courbe de précision-rappel
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
plt.plot(recall, precision, marker='.')
plt.title('Courbe de Précision-Rappel')
plt.xlabel('Rappel')
plt.ylabel('Précision')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement des données
data_path = 'data.csv'
data = pd.read_csv(data_path)

# Suppression des colonnes inutiles
if 'Unnamed: 32' in data.columns:
    data.drop(['Unnamed: 32'], axis=1, inplace=True)

# Encodage des étiquettes
label_encoder = LabelEncoder()
data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])

# Séparation en caractéristiques et cible
X = data.drop('diagnosis', axis=1).values
y = data['diagnosis'].values

# Équilibrage avec SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Normalisation des caractéristiques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Division en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42)

# Redimensionnement des données pour LSTM
X_train_lstm = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# Construction du modèle LSTM
model = Sequential([
    LSTM(64, input_shape=(1, X_train.shape[1]), activation='relu', return_sequences=True),
    Dropout(0.2),
    LSTM(32, activation='relu', return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entraînement du modèle
history = model.fit(X_train_lstm, y_train, epochs=20, batch_size=16, validation_split=0.2, verbose=1)

# Évaluation du modèle
loss, accuracy = model.evaluate(X_test_lstm, y_test, verbose=0)
print(f'\nModel Accuracy: {accuracy:.4f}')

# Prédiction sur l'ensemble de test
y_pred_proba = model.predict(X_test_lstm).ravel()
y_pred = np.where(y_pred_proba > 0.5, 1, 0)

# Calcul des métriques
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Affichage des métriques
print(f'\nAccuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'ROC AUC: {roc_auc:.4f}')

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.title('Matrice de confusion')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()

# Courbe ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='r', label='Chance level')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Courbes d'apprentissage pour la précision et la perte
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```
